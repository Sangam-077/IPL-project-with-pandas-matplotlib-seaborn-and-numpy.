```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#import the data
data=pd.read_csv(r"C:\Users\HP\OneDrive\เอกสาร\Pandas project\matches (1).csv")

#looking for null values
data.isnull().sum()
```

```python
#filling the null values with n/a 
data.fillna("n/a", inplace=True)
```

```python

#having a glance at the first five records of the dataset
data.head()
```

```python

#Lookin at the number of rows and columns in the dataset
data.shape
```

```python
#Getting the frequency of most matches won by any team
winning_team=data['winner'].value_counts()
successful_team=winning_team.idxmax() #team with most wins 
most_successful_team_wins=winning_team.max() #number of games the  successful team won

print(f"the most successful team  is {successful_team} with {most_successful_team_wins} wins ")
```

```python
#plotting thee win counts of all team
plt.figure(figsize=(10,6))
sns.barplot(x=winning_team.index,y=winning_team.values)
plt.title("Most successful teams")
plt.xlabel("Teams")
plt.ylabel("number of wins")
plt.xticks(rotation=90)
plt.show()
```

```python
#Getting the top 5 players with most man of the match awards
top_5_MoM=data["player_of_match"].value_counts()[0:5]
print(top_5_MoM)
```

```python
#making a bar-plot for the top 5 players with most man of the match awards
plt.Figure(figsize=(10,6))
sns.barplot(x=top_5_MoM.index, y=top_5_MoM.values)
plt.title("Top 5 Players with Most 'Man of the Match' Awards")
plt.xlabel("Player")
plt.ylabel("Number of Awards")
plt.xticks(rotation=45)
plt.show()
```

```python
#Getting the frequency of most man of the match awards
MoM=data['player_of_match'].value_counts()[0:10]
Most_MoM=MoM.idxmax()
number_of_MoM_received_by_a_single_player=MoM.max()

print(f"The most man of the match received by any player is {Most_MoM} received {number_of_MoM_received_by_a_single_player} times")
```

```python
#getting the frequency of result either won,drawn or lose
data['result'].value_counts()
```

```python
#Finding out the number of toss wins w.r.t each team
data['toss_winner'].value_counts()
```

```python
#Extracting the records where a team won batting first
batting_first=data[data["win_by_runs"]!=0]
```

```python
#Looking at the head
batting_first.head()
```

```python
#Making a histogram
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'])

plt.title("teams won by runs")
plt.xlabel("teams")
plt.ylabel("wins")
plt.show()
```

```python
#teams with most wins after batting first
batting_first['winner'].value_counts()
```

```python
#Making a bar-plot for top 3 teams with most wins after batting first
plt.figure(figsize=(7,7))
batting_first_winner=batting_first['winner'].value_counts()[0:3]
plt.bar(batting_first_winner.index,batting_first_winner.values, color=["blue","yellow","red"])
plt.show()  

```

```python
# representing the data in pie-chart
team_wins = batting_first["winner"].value_counts()

# Plotting the pie chart
plt.figure(figsize=(10, 10))
plt.pie(team_wins.values, labels=team_wins.index, autopct='%1.1f%%', startangle=200)
plt.title("Win Distribution of All Teams (Batting First)")
plt.axis('equal')  # Make it a circle
plt.show()
```

```python
#extracting those records where a team has won after batting second
second_batting_winner=data[data["win_by_wickets"]!=0]
second_batting_winner['winner'].value_counts()
```

```python
#plotting them in pie-chart
second_winner=second_batting_winner['winner'].value_counts()
plt.figure(figsize=(10,7))
plt.pie(second_winner.values, labels=second_winner.index, autopct='%1.1f%%', startangle=200)
plt.title("second batting winning team")
plt.axis("equal")
plt.show()


```

```python
#representing in chart
import numpy as np
second_winner=second_batting_winner['winner'].value_counts()
colors=np.random.rand(len(second_winner),3)
plt.figure(figsize=(10,7))
plt.bar(second_winner.index,second_winner.values,color=colors)
plt.title("second batting winning team")
plt.xlabel("teams")
plt.ylabel("percentage of game won")
plt.xticks(rotation=45)
plt.grid(axis="y")

plt.show()

```
